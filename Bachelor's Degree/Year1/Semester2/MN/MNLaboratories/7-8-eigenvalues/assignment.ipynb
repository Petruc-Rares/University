## Problema 1

Implementati funcția ```power_method(A, x0, tol, max_steps)``` ce primește o matrice pătratică $A$, o aproximare ințială a vectorului propriu domninant $x_0$, o toleranță sub care se poate opri algoritmul $tol$ și un număr maxim de pași $max\_steps$ și returnează o valoarea proprie dominantă a matricei $A$, $lambda$, un vector propriu normat asociat valorii proprii $lambda$, $x$, eroarea de la ultimul pas $err$ și numărul de pași efectuați $steps$.

Eroarea de la fiecare pas se calculează în funcție de vectorul propriu aproximat, nu de valoarea proprie aproximată.

````octave
function [lambda, x, err, steps] = power_method(A, x0, tol, max_steps)

    steps = 0;
    prev_x = x0;
    while 1
    
        x = A*prev_x;
        x = x/norm(x);
        lambda = x'*A*x;
        err = norm(x - prev_x);
        
        if ((++steps == max_steps) || (err < tol))
        
            return;
        
        endif
    
        prev_x = x;
    
    endwhile
    
endfunction
````

Eroarea de la fiecare pas se calculează în funcție de vectorul propriu aproximat, nu de valoarea proprie aproximată.

````octave
function [X, l, total_steps] = orthogonal_power_method(A, tol, max_steps)
    
    X = [];
    l = [];
    n = length(A);

    for i = 1:n
    
        prev_x = rand(n, 1);
        total_steps = 0;
        
        while 1
        
            for j = 1:size(X, 2)
            
                prev_x = prev_x - X(:, j)' * prev_x * X(:, j);
            
            endfor
            
            x = A*prev_x;
            x = x/norm(x);
            
            lambda = x'*A*x;
            err = norm(x - prev_x);
            
            if ((++total_steps == max_steps) || (err < tol))
            
                l = [l, lambda];
                X = [X, x];
                break;
            
            endif
        
            prev_x = x;
        
        endwhile
    
    endfor

endfunction
````

## Problema 7

Implementati funcția ```hessenberg(A)``` ce primește o matrice $A$ și returnează o matrice tridiagonală cu aceleași valori proprii ca $A$ dacă $A$ este simetrică sau o matrice superior Hessenberg ce aceleași valori proprii ca $A$ dacă a nu este simetrică.

HINT: Trebuie implementat fix același algoritm pentru ambele cazuri, nimic diferit în cod.

````octave
function [R_] = hessenberg(A)

    R_ = 1;
    n = length(A);

        
    for k = 1:n-2

        w = zeros(n, 1);
        alpha = -sign(A(k+1, k))*norm(A(k+1:n, k));
        r = sqrt(1/2*alpha^2 - 1/2*alpha*A(k+1, k));

        w(k+1) = (A(k+1, k) - alpha)/2/r;

        for j = k+2:n

            w(j) = A(j, k)/2/r;

        endfor

        P = eye(n) - 2*w*w';

        A = P*A*P;

    endfor
    
    R_ = A;

end
````

````octave
function [X, l, steps] = qr_algorithm(A, max_steps)

    X = [];
    l = [];
    steps = 0;
    prev_A = A;
    X = eye(length(A));
    
   while 1
    
        [Q, R] = qr(prev_A);
        X = X*Q;
        A = R * Q;

        if ((++steps == max_steps) || (diag(A) == diag(prev_A)))

            l = diag(A)';
            return;


        endif

        prev_A = A;

    endwhile

endfunction
````

````octave
function [P] = getMatrRotatie(A, i ,j, n)

    % il facem pe A(i, j) = 0
    % sn = sin
    % cs = cos
    sn = -A(i, j)/sqrt(A(j, j)^2 + A(i, j)^2);
    cs = A(j, j)/sqrt(A(j, j)^2 + A(i, j)^2);
    
    P = eye(n);
    
    P(j, j) =  P(i, i)= cs;
    P(i, j) = sn;
    P(j, i) = -sn;
    

endfunction
````

````octave
A = [1441 12 0 0 0; 12 293 943 0 0; 0 943 451 9 0; 0 0 9 41 2; 0 0 0 2 131];
valPropriiReale = eig(A);

% dimensiunea matricei A
n = length(A);

valPropriiAprox = [];

% incepem de la P2 (coresp (2, 1))


for i = n:-1:2

    while (abs(A(i, i-1)) > 1e-3)

        j = 1;
        Q = eye(i);
        for k = 2:i

                P = getMatrRotatie(A, k, j, i);
                A = P * A;
                Q = Q * P';


                j++;

        endfor

        
        R = A;

        A = R * Q;

    endwhile
    
    A;

    valPropriiAprox = [valPropriiAprox, A(i, i)];

    A = A(1:i-1, 1:i-1);
    

endfor

% in final ramanem doar cu A(1, 1)
valPropriiAprox = [valPropriiAprox, A(1, 1)];

disp(valPropriiAprox);
disp(valPropriiReale');
norm(sort(valPropriiAprox)-sort(valPropriiReale'))
````

````octave
function [minSigma] = dmin(a, b, pct)

    minValue = a-pct;
    minSigma = a;
    if (abs(b-pct) < abs(minValue))

        minSigma = b;
        minValue = b-pct;
    
    endif

endfunction
````

````octave
A = [3 1 0; 1 3 1; 0 1 3];

valPropriiReale = eig(A);

% dimensiunea matricei A
n = length(A);

valPropriiAprox = [];

sigmauri = 0;

for i = n:-1:3

    A = A(1:i, 1:i);

    val = 0;

    while (abs(A(i, i-1)) > 1e-8)
    
        valprop = eig(A(i-1:i, i-1:i));
        sigma = dmin(valprop(1), valprop(2), A(i, i));
        
        val = val+sigma;
    
        A = A - sigma * eye(i);

        j = 1;
        Q = eye(i);
        for k = 2:i

                P = getMatrRotatie(A, k, j, i);
                A = P * A;
                Q = Q * P';

                j++;

        endfor

        
        R = A;

        A = R * Q;


    endwhile

    break;
    val = val + A(i, i) + sigmauri;
    
    sigmauri = val - A(i, i);

    valPropriiAprox = [valPropriiAprox, val];

    break;

endfor
A

A = A(1:2, 1:2);

valPropriiAprox = [valPropriiAprox, sigmauri + eig(A)'];

disp(sort(valPropriiAprox));
disp(sort(valPropriiReale'));
norm(sort(valPropriiAprox)-sort(valPropriiReale'))
````
